```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Prediction using voltage, current, and temperature

```python
real = pd.read_csv("../DATA/DataFrame/B0005_capacity_df.csv")
MC_LSTM = pd.read_csv("./MC_LSTM.csv")
MC_AttentionLSTM = pd.read_csv("./MC_AttentionLSTM.csv")
MC_2MAttn = pd.read_csv("./MC_2MAttn.csv")
MC_PE_2MAttn = pd.read_csv("./MC_PE_2MAttn.csv")
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.update_layout(template='plotly_white')

fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=real["capacity"], mode='lines',name='Real'))
fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=MC_LSTM.iloc[:,0], mode='lines',name='LSTM', line_dash='dash'))
fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=MC_AttentionLSTM.iloc[:,0], mode='lines+markers',name='Attention_LSTM', line_dash='dot'))
fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=MC_2MAttn.iloc[:,0], mode='lines+markers',name='No_positional_encoding', marker_symbol='star'))
fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=MC_PE_2MAttn.iloc[:,0], mode='lines+markers',name='Proposed Model', marker_symbol='arrow-right'))
fig.add_trace(go.Scatter(x=list(range(len(real.index))), y=[1.4 for i in range(len(real.index))], mode='lines',name='Threshold', line_color='black'))

fig.update_layout(width=1300,height=850)
fig.update_xaxes(range=[0, 166])
fig.update_yaxes(range=[1.2, 2])

fig.update_xaxes(title_text='Cycle')
fig.update_yaxes(title_text='Capacity(Ah)')
fig.update_xaxes(title_font_size =20,
                title_font_color='black',
                title_font_family='Courier')
fig.update_yaxes(title_font_size =20,
                title_font_color='black',
                title_font_family='Courier')

fig.update_layout(title_text="Prediction Of B0005")
fig.update_layout(
                title_x = 0.5,
                title_y = 0.9,
                title_xanchor = "center",
                title_yanchor = "middle",
                title_font_size =30,
                title_font_color='black',
                title_font_family='Courier')

fig.update_layout(
    legend_yanchor="top",
    legend_xanchor="right",
    legend_font_size=23
)

fig.show()
```
