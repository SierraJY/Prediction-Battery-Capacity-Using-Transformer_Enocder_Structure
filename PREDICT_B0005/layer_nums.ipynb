```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Multi-head Self Attention Layer 개수별 성능

```python
plt.rcParams['figure.figsize'] = [8, 6]
plt.xlim([1, 6])

MAPEs_b0005 = [2.582, 1.252, 1.676, 1.86, 2.952, 3.437]
plt.plot(range(1,7), MAPEs_b0005, label="MAPE",linewidth=2, color='red')
plt.title("MAPE Performance by Number of Multi-Head Self-Attention Layers Used")
plt.xlabel(xlabel='Number of Multi-Head Self-Attention Layers', labelpad=15)
plt.ylabel(ylabel='MAPE', labelpad=15)
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = go.Figure()
fig.update_layout(template='plotly_white')

fig.add_trace(go.Scatter(x=list(range(1,7)), y=MAPEs_b0005, name="MAPE", mode='lines+markers'))


fig.update_layout(width=1100,height=800)
fig.update_xaxes(range=[0.9, 6.2])
fig.update_yaxes(range=[1.2, 3.5])

fig.update_xaxes(title_text='Cycle')
fig.update_yaxes(title_text='Capacity')

fig.update_traces(marker_color= 'red', marker_size=15,)
fig.update_xaxes(title_text='Number of Multi-Head Self Attention Layers')
fig.update_yaxes(title_text='MAPE')
fig.update_xaxes(title_font_size =30,
                title_font_color='black',
                title_font_family='Courier')
fig.update_yaxes(title_font_size =30,
                title_font_color='black',
                title_font_family='Courier')

fig.show()
```
